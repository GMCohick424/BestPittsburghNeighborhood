```python
import pandas as pd
import matplotlib.pyplot as plt
import operator#sory items by value

empy = pd.read_csv("../Datasets/Traffic_Signs.csv")#reads vile csv.csv 
neighborhood = {}#stores amount of stop signs for each neighborhood
for index, row in empy.iterrows():
    if row["neighborhood"] in neighborhood:
        neighborhood[row["neighborhood"]] += 1
    else:
        neighborhood[row["neighborhood"]] = 1
        
highest = dict(sorted(neighborhood.items(), key=operator.itemgetter(1), reverse=True))

plt.bar(list(neighborhood.keys())[:10], list(neighborhood.values())[:10])#creates plot bar with top 10 neighborhoods
plt.xticks(rotation=90)# rotates 90 degrees
```

```python
import pandas as pd
from tabulate import tabulate# helps print table

stop_sign_data = pd.read_csv("../Datasets/Traffic_Signs.csv")

stop_sign_ahead_data = stop_sign_data[stop_sign_data['description'] == "Stop Sign Ahead"]

stop_sign_per_neighborhood = stop_sign_ahead_data['neighborhood'].value_counts().head(10)

stop_sign_table = pd.DataFrame({
    'Neighborhood': stop_sign_per_neighborhood.index,
    'Stop Sign Ahead Count': stop_sign_per_neighborhood.values
})

print(tabulate(stop_sign_table, headers='keys', tablefmt='fancy_grid'))
```
