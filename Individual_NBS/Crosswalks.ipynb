```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../Datasets/Crosswalks.csv')
num = df['neighborhood'].value_counts()
# brookline = df[df['neighborhood'] == 'Brookline'].shape[0]
under20 = num[num < 20]
print(num)
print(under20)

plt.figure(figsize=(8, 6))
under20.plot(kind='bar')
plt.xlabel('neighborhood')
plt.ylim(0, 18)
plt.ylabel('Count')
plt.title(f'Neighborhoods with less than 20 crosswalks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Conclusion

We find that the best neighborhoods in Pittsburgh are those with the fewest crosswalks, as this allows for fewer interruptions to continuous driving. With the dataset showing neighborhoods with fewer than 20 crosswalks, the neighborhoods with the least pedestrian interferenceand thus the most "speed-friendly" roads are New Homestead, St. Clair, and Chartiers City, which each have minimal crosswalk presence.

This lack of crosswalks means these areas are less likely to have frequent stops, making them ideal for uninterrupted driving. Conversely, neighborhoods like Central Business District with 402 crosswalks, South Side Flats with 262 crosswalks, and Squirrel Hill South with 239 crosswalks, present more potential obstacles, decreasing drivers' ability to speed.
